```python
gamma = 1.4

#Este t es el t final hasta el que queremos iterar, viene como entrada en el archivo de C
t = 0.1

ite = 1000
dt = t/ite #1E-4

tam = 1000 
dx = 20.0/(float(tam)) #2E-2

#inicializo una matriz, que en cada entrada tiene una tripleta:= (velocidad, densidad, presion)
matrix = zeros((ite,tam,3))

# Velocidad = matrix[i,j,0]
# densidad = matrix[i,j, 1]
# presion = matrix[i,j,2]

#Condiciones iniciales

for i in range(tam):
    if i<=499:                   #gas de la izquierda----------
        matrix[i,0,1]=1.0        #densidad 1kg/m³
        matrix[i,0,2]=100000.0   #presion 1E5 N/m²
    elif i>499:                  #gas de la derecha------------
        matrix[i,0,1]=0.125      #densidad 0.125kg/m³
        matrix[i,0,2]=10000.0    #presion 1E4 N/m²
        
# "i" eje posicion, "j" eje temporal

#Esta funcion retorna la energia total
def et(i,j):
    return matrix[i,j,2]/(matrix[i,j,1]*(gamma-1.0)) + (matrix[i,j,0]**2)/2.0 

#Esta funcion retorna el vector U, en funcion de cada posicion en la matriz
def Uv(i,j):
    U = zeros(3)
    U[0] = matrix[i,j,1]
    U[1] = matrix[i,j,1]*matrix[i,j,0]
    U[2] = matrix[i,j,1]*et(i,j)
    return U

#Esta funcion retorna el vector F, en funcion de un arreglo (que mas tarde será U para algunas posiciones)
def Fv(arreglo):
    F = zeros(3)
    a = arreglo[0]
    b = arreglo[1]
    c = arreglo[2]
    F[0] = b
    F[1] = (b**2)/a + (gamma-1.0)*(c - (b**2)/(2.0*a))
    F[2] = (b/a)*c + (gamma-1.0)*c*(b/a) - (gamma-1.0)*(b/a)*(b**2)/(2.0*a)
    return F

    
#Esta funcion hace el metodo para una posicion, la cual toma en cuenta la posicion anterior y la sgt en el eje
#espacial, y genera valores para vel, densidad y presion en la sgt posicion en el eje temporal (de una las guarda en la matrix)
def metodo(i,j):
    utemppos = zeros(3)
    utempneg = zeros(3)
    ftemppos = zeros(3)
    ftempneg = zeros(3)
    unew = zeros(3)
    if i==0:
        for h in range(3):
            utemppos[h] = 0.5*( Uv(i+1,j)[h]+Uv(i,j)[h] - (dt/dx)*( Fv( Uv(i+1,j) )[h] - Fv(Uv(i,j))[h] ) )
            #Si esta en el borde de la matriz, hace las cosas con las condiciones iniciales
            utempneg[h] = 0.5*( Uv(0,0)[h]+Uv(i,j)[h] - (dt/dx)*( Fv( Uv(0,0) )[h] - Fv(Uv(i,j))[h] ) )
        ftemppos = Fv(utemppos)
        ftempneg = Fv(utempneg)    
    if i==(tam-1):
        for h in range(3):
            #Si esta en el borde de la matriz, hace las cosas con las condiciones iniciales
            utemppos[h] = 0.5*( Uv(tam-1,0)[h]+Uv(i,j)[h] - (dt/dx)*( Fv( Uv(tam-1,0) )[h] - Fv(Uv(i,j))[h] ) )
            utempneg[h] = 0.5*( Uv(i-1,j)[h]+Uv(i,j)[h] - (dt/dx)*( Fv( Uv(i-1,j) )[h] - Fv(Uv(i,j))[h] ) )
        ftemppos = Fv(utemppos)
        ftempneg = Fv(utempneg)
    else:
    #En este caso no hay de que preocuparse, puesto que si existen valores para "i-1" e "i+1"
        for h in range(3):
            utemppos[h] = 0.5*( Uv(i+1,j)[h]+Uv(i,j)[h] - (dt/dx)*( Fv( Uv(i+1,j) )[h] - Fv(Uv(i,j))[h] ) )
            utempneg[h] = 0.5*( Uv(i-1,j)[h]+Uv(i,j)[h] - (dt/dx)*( Fv( Uv(i-1,j) )[h] - Fv(Uv(i,j))[h] ) )
        ftemppos = Fv(utemppos)
        ftempneg = Fv(utempneg)
    #Se generan y guardan los valores vel, densidad y presion para el siguiente paso en el tiempo
    for h in range(3):
        unew[h] = Uv(i,j)[h] - (dt/dx)*( ftemppos[h] - ftempneg[h] )
    matrix[i, j+1, 1] = unew[0]
    matrix[i, j+1, 0] = unew[1]/unew[0]
    matrix[i, j+1, 2] = (gamma-1.0)*( unew[2] - ( unew[0]*(unew[1]/unew[0])**2 )/2.0)
```

```python
#Ahora vamos a implementar el metodo, recorriendo toda la matriz, en i de "0" a "tam-1", y en j de "0" a "ite-2"
#puesto que si lo hacemos de "0" a "ite-1", el metodo no tendria en donde guardar lo que genera, puesto que la matriz se
#acaba en j=ite-1

for i in range(tam):
    for j in range(ite-1):
        if i % 100 and j % 100:
            print "Va en la entrada " + str(i)+"\t"+str(j)+" de la matriz"
        metodo(i,j)
        
#Ya teniendo la matriz hecha, con todos los valores de velocidad, densidad y presion, para cada tiempo "j" y posicion "i"
#Podemos hacer el archivo de texto con las 4 columnas, para el ultimo tiempo (j=ite-1)
# POSICION    VELOCIDAD    PRESION    DENSIDAD

file = open("estado_"+str(t)+".dat", "w")

for i in range(tam):
    file.write( str(-10+(i*20.0)/999.0)+"\t"+str(matrix[i,ite-1,0])+"\t"+str(matrix[i,ite-1,2])+"\t"+str(matrix[i,ite-1,1])+"\n")

file.close()
   
```
